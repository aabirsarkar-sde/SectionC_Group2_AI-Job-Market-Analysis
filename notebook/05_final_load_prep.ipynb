# Initial Setup and Data Loading

## Data Analysis from PDF Content

```python
import pandas as pd
import numpy as np

# LOAD DATA
df = pd.read_csv("/content/ai_job_trends_dataset.csv")
```

### Column Standardization

```python
df = df.rename(columns={
    'Projected Openings (2030)': 'openings_2030',
    'Job Openings (2024)': 'openings_2024',
    'Median Salary (USD)': 'salary',
    'Automation Risk (%)': 'automation_risk',
    'Experience Required (Years)': 'experience',
    'Remote Work Ratio (%)': 'remote_ratio',
    'Required Education': 'education',
    'Job Title': 'job_title',
    'Industry': 'industry',
    'AI Impact Level': 'ai_impact'
})
```

### Feature Engineering

```python
df['growth_rate'] = (df['openings_2030'] - df['openings_2024']) / df['openings_2024']
df['risk_category'] = pd.cut(df['automation_risk'], bins=[0,33,66,100],
                            labels=['Low','Medium','High'])
df['growth_category'] = pd.cut(df['growth_rate'], bins=[-1,0,0.2,1],
                              labels=['Declining','Moderate','High'])
```

### KPI Table

```python
kpi_df = pd.DataFrame({
    "Metric": ["Total Jobs","Avg Salary","Median Salary","Avg Growth","Avg Risk"],
    "Value": [
        df.shape[0],
        df['salary'].mean(),
        df['salary'].median(),
        df['growth_rate'].mean(),
        df['automation_risk'].mean()
    ]
})
```

### Dashboard Data Preparation

```python
# DASHBOARD 1: JOB GROWTH
industry_summary = df.groupby('industry').agg({
    'growth_rate':'mean',
    'job_title':'count'
}).reset_index()
job_summary = df.groupby('job_title').agg({
    'growth_rate':'mean'
}).reset_index()

# DASHBOARD 2: SALARY
salary_summary = df.groupby(['industry','ai_impact']).agg({
    'salary':'mean'
}).reset_index()

# DASHBOARD 3: AUTOMATION RISK
risk_summary = df.groupby(['risk_category','industry']).size().reset_index(name='count')

# DASHBOARD 4: WORKFORCE
workforce_summary = df.groupby('education').agg({
    'salary':'mean',
    'growth_rate':'mean',
    'experience':'mean'
}).reset_index()
remote_summary = df.groupby('remote_ratio').agg({
    'salary':'mean',
    'growth_rate':'mean'
}).reset_index()
```

### Save Files

```python
kpi_df.to_csv("/content/kpi_data.csv", index=False)
industry_summary.to_csv("/content/industry_summary.csv", index=False)
job_summary.to_csv("/content/job_summary.csv", index=False)
salary_summary.to_csv("/content/salary_summary.csv", index=False)
risk_summary.to_csv("/content/risk_summary.csv", index=False)
workforce_summary.to_csv("/content/workforce_summary.csv", index=False)
remote_summary.to_csv("/content/remote_summary.csv", index=False)
df.to_csv("/content/final_dashboard_data.csv", index=False)
```
